```python
import pandas as pd
```

```python
df = pd.read_csv('stft_5_97020.csv', index_col=0)
```

```python
feature_list=['chroma_0','chroma_1','chroma_2','chroma_3','chroma_4']
```

```python
df
```

```python
def get_audio_features_slices(df, features, interval=10, overlap=2):
    slices = []
    end = df.shape[0]
    i = 0
    while True:
        if i+10 >= df.shape[0]:
            break
            # Alternative? slice_temp = df.iloc[i:-1][features]
            slice_temp = df.iloc[-10:][features]
            slices.append(slice_temp)
            
        else:
            slice_temp = df.iloc[i:i+10][features]
            slices.append(slice_temp.mean(axis=0))
            i+=interval-overlap
    return slices
```

```python
slices = get_audio_features_slices(df, feature_list, overlap=8)
```

```python
slices
```

```python
pd.concat(slices).reset_index(drop=True).to_csv('audio_feature_slices.csv')
```

```python
for i in range(len(slices)):
    slices[i].reset_index(drop=True).to_csv('audio_feature_slices_'+str(i+1)+'.csv')
```
