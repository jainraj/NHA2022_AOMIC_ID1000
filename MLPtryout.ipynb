```python
PhenoDf_IQ = pd.DataFrame(PhenoDf[['participant_id', 'IST_intelligence_total']])
```

```python
inputDf = pd.merge(ConnectDf, PhenoDf_IQ, on = "participant_id")
```

```python
# separate the target feature in an array `y` and drop it from the rest of the data `X`
y = inputDf['IST_intelligence_total']
X = inputDf.drop(['IST_intelligence_total'] , axis = 1)
labels = X.columns

# test and training set split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=42)
X_train.shape
```

```python
X_train = X_train.drop(['Unnamed: 0', 'participant_id'], axis = 1)
X_test = X_test.drop(['Unnamed: 0', 'participant_id'], axis = 1)
```

```python
scaler = MinMaxScaler()
y_train = scaler.fit_transform(y_train.values.reshape(-1, 1)).ravel()
y_test = scaler.transform(y_test.values.reshape(-1, 1)).ravel()
```

```python
LR = LinearRegression()
LR.fit(X_train, y_train)
y_pred_LR = LR.predict(X_test)
LR.coef_
```

```python
mse_kq = mean_squared_error(y_pred_LR, y_test)
mse_kq
```

```python
ridge = Ridge(alpha=1)
ridge.fit(X_train, y_train)
y_pred_ridge = ridge.predict(X_test)
ridge.coef_
```

```python
mse_kq = mean_squared_error(y_pred_ridge, y_test)
mse_kq
```

```python
## Grid search with cross validation

parameters = {'alpha':[0.1, 10]}

# define the model/ estimator
model = Ridge()

# define the grid search
Ridge_reg= GridSearchCV(model, parameters, scoring='neg_mean_squared_error',cv=5)

#fit the grid search
Ridge_reg.fit(X_train, y_train)

# best estimator
print(Ridge_reg.best_estimator_)

# best model
best_model = Ridge_reg.best_estimator_
best_model.fit(X_train, y_train)
```

```python
y_pred_ridge = best_model.predict(X_test)
# best_model.coef_
mse_kq = mean_squared_error(y_pred_ridge, y_test)
mse_kq
```
